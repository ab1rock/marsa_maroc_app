```python
import cv2
import supervision as sv
from ultralytics import YOLO

# Path to your image file
image_path = "./Container-Code-Detection-1/test/images/Container-148_jpg.rf.5bd0c014e0665b33a344c02d2fd19a98.jpg"  #change image_path

# Load the YOLO model
try:
    model = YOLO('./pre_trained_on_roboflow_dataset.pt')  # Replace with the path to your YOLO model
    model.to('cuda')  # Assurez-vous que le modèle est chargé sur le GPU
except AttributeError as e:
    print("Error loading model:", e)

# Initialize annotators
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# Load the image
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print(f"Error: Unable to load image at {image_path}")
else:
    # Resize the image to reduce image size
    resized_image = cv2.resize(image, (640, 480))  # Specify the desired width and height
    
    # Process the image for detection
    try:
        results = model(resized_image, device='cuda')[0]  # Utilise CUDA pour l'inférence
    except Exception as e:
        print("Error during model prediction:", e)
    
    # Convert YOLO results to detections
    detections = sv.Detections.from_ultralytics(results)
    
    # Annotate the image
    annotated_image = box_annotator.annotate(scene=resized_image, detections=detections)
    
    # Save cropped images based on bounding boxes with enlarged margins
    margin = 0  # Define the margin to be added to each side of the box
    for i, box in enumerate(detections.xyxy):
        x1, y1, x2, y2 = map(int, box[:4])  # Extract bounding box coordinates

        # Add margin and ensure coordinates are within image bounds
        x1 = max(0, x1 - margin)
        y1 = max(0, y1 - margin)
        x2 = min(resized_image.shape[1], x2 + margin)
        y2 = min(resized_image.shape[0], y2 + margin)

        cropped_image = resized_image[y1:y2, x1:x2]  # Crop the image based on the box coordinates with margin
        cv2.imwrite(f"cropped_image_{i}.jpg", cropped_image)  # Save the cropped image
    
    # Display the annotated image using OpenCV
    cv2.imshow("Annotated Image", annotated_image)
    
    # Press any key to close the window
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Optionally, save the annotated image
    cv2.imwrite("annotated_image.jpg", annotated_image)
```

```python
import cv2
import easyocr
import torch
from datetime import datetime

# Check if CUDA is available
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Load the image
img_path = 'restcrud/media/uploads/cropped_image_b89c_0.jpg'  # Replace with the path to your image
img = cv2.imread(img_path)

# Convert to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Save the processed image if necessary
cv2.imwrite('processed_image.jpg', img_gray)

# Apply EasyOCR
reader = easyocr.Reader(['en'], gpu=torch.cuda.is_available())  # Load the model for English with GPU if available
result = reader.readtext(img_gray)

# Combine all detected text into a single line and calculate the average probability
combined_text = ' '.join([text for _, text, _ in result])
average_probability = sum([prob for _, _, prob in result]) / len(result) if result else 0

# Get the current time for the detection
detection_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

# Process the combined_text to apply corrections (e.g., changing 'I' to '1', etc.)


fixed_code = list(combined_text)

# Iterate through each character starting from the 5th character (index 4)
for i in range(4,len(fixed_code)):
    if fixed_code[i] == 'I':
        fixed_code[i] = '1'
    elif fixed_code[i] == 'A':
        fixed_code[i] = '4'
    elif fixed_code[i] == 'G':
        fixed_code[i] = '6'
    elif fixed_code[i] == 'B':
        fixed_code[i] = '8'
    elif fixed_code[i] == 'J':
        fixed_code[i] = '9'
    elif not (fixed_code[i].isdigit() or fixed_code[i].isalpha()):
        fixed_code[i] = ''
        

# Join the fixed_code list back into a string
corrected_code = ''.join(fixed_code)
corrected_code=corrected_code[:11]
# Print the combined detected text, average probability, detection time, and corrected code
print(f"Combined detected text: {combined_text}")
print(f"Average Probability: {average_probability:.2f}")
print(f"Detection Time: {detection_time}")
print(f"Corrected Code: {corrected_code}")

# Display the grayscale image

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### VERSION VIDEO ( OBJECT DETECTION)

```python
import cv2
import supervision as sv
from ultralytics import YOLO

# URL of your video stream
video_url = "http://192.168.1.6:8080/video"

# Load the YOLO model
try:
    model = YOLO('./pre_trained_on_roboflow_dataset.pt')  # Replace with the path to your YOLO model
except AttributeError as e:
    print("Error loading model:", e)

# Initialize annotators
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# Open the video stream
cap = cv2.VideoCapture(video_url)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Resize the frame to reduce image size
    resized_frame = cv2.resize(frame, (640, 480))  # Specify the desired width and height
    
    # Process frame for detection
    try:
        results = model(resized_frame)[0]
    except Exception as e:
        print("Error during model prediction:", e)
        continue

    detections = sv.Detections.from_ultralytics(results)
    annotated_image = box_annotator.annotate(scene=resized_frame, detections=detections)
    
    # Display the annotated frame using OpenCV
    cv2.imshow("Annotated Video Stream", annotated_image)
    
    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
import supervision as sv
from ultralytics import YOLO

# Chemin vers le dossier contenant les images
input_folder = "./Container-Code-Detection-1/test/images/"
output_folder = "./annotated_images"

# Créer le dossier de sortie s'il n'existe pas
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Charger le modèle YOLO
try:
    model = YOLO('./pre_trained_on_roboflow_dataset.pt')  # Remplacez par le chemin vers votre modèle YOLO
except AttributeError as e:
    print("Erreur lors du chargement du modèle:", e)

# Initialiser les annotateurs
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# Parcourir toutes les images dans le dossier d'entrée
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)

        # Vérifier si l'image a été chargée avec succès
        if image is None:
            print(f"Erreur: Impossible de charger l'image à {image_path}")
            continue

        # Redimensionner l'image pour réduire la taille
        resized_image = cv2.resize(image, (640, 480))  # Spécifiez la largeur et la hauteur souhaitées

        # Traiter l'image pour la détection
        try:
            results = model(resized_image)[0]
        except Exception as e:
            print("Erreur lors de la prédiction du modèle:", e)
            continue

        # Convertir les résultats YOLO en détections
        detections = sv.Detections.from_ultralytics(results)

        # Annoter l'image
        annotated_image = box_annotator.annotate(scene=resized_image, detections=detections)

        # Chemin de sauvegarde de l'image annotée
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, annotated_image)
        print(f"Image annotée sauvegardée à {output_path}")
```
